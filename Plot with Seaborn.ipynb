# Plot with Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips") #step1
sns.set_style("whitegrid") #step2
```

```python
g = sns.lmplot(x = "tip", y = "total_bill", data = tips, aspect = 2) #step3
g = (g.set_axis_labels("Tip", "Total bill(USD)").
set(xlim = (0,10), ylim = (0, 100))) #strep4
plt.title("title")
plt.show(g)
```

## 1. Data

```python
import pandas as pd
import numpy as np
uniform_data = np.random.rand(10, 12)
data = pd.DataFrame({"x":np.arange(1, 101),
                    "y":np.random.normal(0,4, 100)})
```

```python
titanic = sns.load_dataset("titanic")
iris = sns.load_dataset("iris")
```

## 2. Figure Aesthetics
### Seaborn styles

```python
f, ax = plt.subplots(figsize=(5, 6))
sns.set()
sns.set_style("whitegrid")
sns.set_style("ticks", 
              {"xtick.major.size":8,
              "ytick.major.size":8})
sns.axes_style("whitegrid")
```

### Context Functions

```python
sns.set_context("talk")
sns.set_context("notebook", font_scale = 1.5,
               rc = {"lines.linewidth":2.5})
```

### Color Palette

```python
sns.set_palette("husl", 3)
sns.color_palette("husl")
flatui = ["#9b59b6","#3498db","#95a5a6","#e74c3c","#34495e","#2ecc71"]
sns.set_palette(flatui)
```

## 3. Plotting with Seaborn
### Axis Grids

```python
g = sns.FacetGrid(titanic, col = "survived", row = "sex")
g = g.map(plt.hist, "age")
sns.factorplot(x = "pclass", y = "survived", hue = "sex", data = titanic)
```

```python
sns.lmplot(x= "sepal_width", y = "sepal_length", hue = "species", data = iris)
```

```python
h = sns.PairGrid(iris)
h = h.map(plt.scatter)
sns.pairplot(iris)
```

```python
i = sns.JointGrid(x = "x", y = "y", data = data)
i = i.plot(sns.regplot, sns.distplot)
sns.jointplot("sepal_length", "sepal_width", data = iris, kind = "kde")
```

### Categorical Plots
#### Scatterplot

```python
sns.stripplot(x = "species", y = "sepal_length",  data = iris)
```

```python
sns.swarmplot(x = "species", y = "petal_length", data = iris)
```

#### Bar Chart

```python
sns.barplot(x = "sex", y = "survived", hue = "class", data = titanic)
```

#### Count Plot

```python
sns.countplot(x = "deck", data = titanic, palette = "Greens_d")
```

#### Point Plot

```python
sns.pointplot(x ="class", y = "survived", hue = "sex", data = titanic,
          palette ={"male":"g", "female":"m"}, markers = ["^", "o"],
         linestytles = ["-", "--"])
```

#### Box Plot

```python
sns.boxplot(x = "alive", y = "age", hue = "adult_male", data = titanic)
```

```python
sns.boxplot(data = iris, orient = "h")
```

#### Violinplot

```python
sns.violinplot(x = "age", y = "sex", hue = "survived", data = titanic)
```

### Regression Plots

```python
sns.regplot(x= "sepal_width", y = "sepal_length", data = iris, ax = ax)
```

### Distribution Plots

```python
sns.distplot(data.y, kde = False, color = "b")
```

### Matrix Plots

```python
sns.heatmap(uniform_data, vmin = 0, vmax = 1)
```
